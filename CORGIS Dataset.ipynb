# CORGIS Weather Dataset Analysis
By analying this dataset, we can obtain a general idea of the trends of specific weather traits.
## Structure
Each dictionary in the dataset represents the data collected by one station. Consider the following station:

```python
%run analysis.py structure 1
```

## Analysis
For a given number of stations, we can plot a feature of interest against time.

Consider the following example:

```python
%run analysis.py plot 7000 Avg Temp
```

The red line is a linear line of best fit. While the temperatures depicted above seem to increase over time, other traits do not. E.g.:

```python
%run analysis.py plot 7000 High Gust
```

We can also plot stations based on their latitudes:

```python
%run analysis.py latitude 20 Min Temp 45 2
```
